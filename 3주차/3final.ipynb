```python
%%writefile src/read_it.py

import os
import requests
import urllib
import mylib # NO! src.mylib

def doIt():
    keyPath=os.path.join(os.getcwd(), 'src', 'key.properties')
    key=mylib.getKey(keyPath)
    KEY=str(key['dataseoul'])
    TYPE='json'
    SERVICE='VwsmTrdhlNcmCnsmpQq'
    
    _url='http://openAPI.seoul.go.kr:8088'
    _key=str(key['dataseoul'])
    _type='json'
    _service='VwsmTrdhlNcmCnsmpQq'
    _start_index=1
    _end_index=1000
    _use_dt=str(2020)
    
    while True:
        _api="http://openapi.seoul.go.kr:8088/495a6e694e6e696338326567777766/json/VwsmTrdhlNcmCnsmpQq/1/1000/2020"
        r=requests.get(_api)
        passenger=r.json()
        if(_start_index==1):
            list_total_count=passenger['VwsmTrdhlNcmCnsmpQq']['list_total_count']
#             print("- Total Count: ", list_total_count)
        for e in passenger['VwsmTrdhlNcmCnsmpQq']['row']:
            print ("{0:10s}\t{1:15s}\t{2:3f}\t{3:2f}".format(e['TRDAR_CD_NM'], e['INCOME_SCTN_CD'], e['EXPNDTR_TOTAMT'], e['PLESR_EXPNDTR_TOTAMT']))        
        _start_index+=10
        _end_index+=10
        if(_end_index > list_total_count):
#             print("----- Ending endIndex=",_end_index)
            break  # exit from the while loop

if __name__ == "__main__":
    doIt()
```
